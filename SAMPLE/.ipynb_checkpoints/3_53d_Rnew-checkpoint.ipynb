```python
id = '2.53d'
idR ='2_53d' 
template = r'''\begin{problem}
Случайные величины $X$ и $Y$ независимы и имеют равномерное
распределение на отрезках $[0;((a))]$ и $[0;((b))]$ соответственно. Для случайной величины $Z=\frac{Y}{X}$ найдите: 
1) функцию распределения $F_Z(x)$;
2) плотность распределения $f_Z(x)$ и постройте график плотности;
3) вероятность $\P(((z1))\leqslant Z\leqslant ((z2)))$.
\end{problem}
\begin{solution}
%\folder ((file_name))
1) Функция распределения $F_Z(x)$ имеет вид:
$
F_Z(x)=\left\{
\begin{array}{l}
0, x\leqslant 0;\\
((Q3)), 0\leqslant x\leqslant ((Q1))\approx ((q1a));\\
((Q4)), x\geqslant((Q1));
\end{array}.
\right.
$
2) Плотность распределения $f_Z(x)$ имеет вид:
$
f_Z(x)=\left\{
\begin{array}{l}
0, x<0;\\
((Q5)), 0\leqslant x\leqslant ((Q1))\approx ((q1a));\\
((Q6)), x\geqslant((Q1));
\end{array}.
\right.
$


\begin{figure}[H]
    \includegraphics[width=0.9\textwidth]{((file_name_without_extension))}
\end{figure}


3) вероятность равна:
$
\P(((z1))\leqslant Z\leqslant ((z2)))=
((p)).
$
\end{solution}
'''
```

```python
def get():
    a = rand(1,10)
    b = rand(1,10)
    t=1/1000*rand(1,1000-1)
    s=1+1/1000*rand(1,1000-1)
    z1=t*b/a
    z2=s*b/a
    q1 =Rational(b/a).limit_denominator(1550000)
    q1a=b/a
    q3= lambda x: (a*x)/(2*b)
    q4= lambda x: 1-simplify(b/(2*a*x))
    q5= lambda x: Rational(a/(2*b)).limit_denominator(1550000)
    q6= lambda x: simplify(b/(2*a*x**2))
    p=(2*s-1)/(2*s)-t/2
    return a,b,z1,z2,q1,q1a,q3(x), q4(x), q5(x), q6(x),p
```

```python
def figure(k,A,B,Z2):
    a1=A
    b1=B
    c1=Z2
    f= lambda x :  0 if (x<=0) else ((0<x) and (x<=b1/a1))*(a1/(2*b1)) or (b1/a1<=x)*(b1/(2*a1*x**2)) 
    fig,ax =plt.subplots(figsize=(10, 5))
    plt.tick_params(labelsize = 20)
    #ax.xaxis.set_major_locator(ticker.MultipleLocator(scale/4))
    #ax.patch.set_alpha(a1)
    ax.set_xlim(xmin=-1, xmax=c1+1)
    ax.xaxis.set_major_locator(ticker.MaxNLocator(8))
    ax.xaxis.set_minor_locator(ticker.MaxNLocator(40))
    #ax.yaxis.set_major_locator(ticker.MultipleLocator(round(2.0/(b1-a1)/4,2)))
    ax.yaxis.set_major_locator(ticker.MaxNLocator(6))
    plt.grid(b=True, color='DarkTurquoise', alpha=0.75, linestyle=':', linewidth=1)
    #axis('scaled')
    plt.ylim(0.,a1/(2*b1)+0.05)
    plt.xlim(-0.5,c1+2)
    x1 = np.linspace(-1, c1+2, 1000)  
    y1 = [f(t) for t in x1]  
    #plt.plot(x, y1, lw=5, alpha=1,color='black')  
    #plt.plot(x, y2, 'r--',lw=5, alpha=0.6)  
    plt.plot(x1, y1, 'k-',lw=5) 
    y0=0
    plt.fill_between(x1,y1,y0, color='orange', alpha=0.2)
    #h = plt.hist(T.rvs(size=10000000), bins=200, density=True)
    #plt.scatter(c, T.pdf(c), s=1000, c='r')
    #plt.scatter(c, 0.005, s=1000, c='b')
    #plt.vlines(x=c,ymin=0.0, ymax=0.5, color='b', linestyle='--',lw=4)
    #plt.text(c-0.05, -0.03, r'$c$',fontsize=40, color ='b', fontweight='bold')
    font = font0.copy()
    font.set_style('italic')
    font.set_weight('bold')
    #font.set_size('xx-large')
    #plt.text(c-0.05, -0.03, r'$c$', color ='b',  fontproperties=font,size=40)
    #plt.text(c+0.1, T.pdf(c), r'$f(c)=\frac{2}{b-a}$',fontsize=40, color ='k', fontweight='bold')
    #plt.text((a+b)/4, T.pdf(c), r'$X\sim Tri([a;b];c)\leftrightarrow X=triang\left(\frac{c-a}{b-a}, a, b-a\right)$',fontsize=40, color ='k', fontweight='bold')
    picture_path = Path(Path.cwd().parent, "GEN", "_Base", "Q"+idR[0], idR+str(k)+'.png')
    fig.savefig(picture_path, bbox_inches='tight')
    #plt.close()
```

```python
nvp = 25 #число вариантов параметризации
file_name = "Problem_with_picture"
param_file = Path(Path.cwd().parent, "GEN", "_Base", "Q"+idR[0], file_name+".txt")
ftxt = open(param_file,'w',encoding='utf-8')
```

```python
for vp in range(1,nvp+1):
    atxt = template
    A, B, Z1, Z2, Q1, Q1a, Q3, Q4,Q5, Q6, P= get()
    figure(vp,A,B,Z2)
    dig = 5
    atxt = atxt.replace('((a))',latex(A))
    atxt = atxt.replace('((b))',latex(B))
    atxt = atxt.replace('((z1))',latex(round(Z1,3)).replace('.', ',\!'))
    atxt = atxt.replace('((z2))',latex(round(Z2,3)).replace('.', ',\!'))
    atxt = atxt.replace('((q1a))',latex(round(Q1a,3)).replace('.', ',\!'))
    atxt = atxt.replace('((Q1))',latex(Q1))
    atxt = atxt.replace('((Q3))',latex(Q3))
    atxt = atxt.replace('((Q4))',latex(Q4))  
    atxt = atxt.replace('((Q5))',latex(Q5))  
    atxt = atxt.replace('((Q6))',latex(Q6))
    atxt = atxt.replace('((p))',latex(round(P,dig)).replace('.', ',\!'))
    atxt = atxt.replace('((file_name))',idR+str(vp)+'.png')
    atxt = atxt.replace('((file_name_without_extension))',idR+str(vp))
    idvp = id + str(vp)+'|\n'
    ftxt.write('{}\n'.format(atxt))
ftxt.close()
```
